## 1. Introducion

From an administration point of view, a country can be seen as a collection of different regions or units. That is, a collection of subnational entities which are given some degrees of autonomy [[1]](https://en.wikipedia.org/wiki/Administrative_division). Two different geocodes standards, or systems, are used and internationally recongnized: **NUTS**, and **GADM**.

Since one may be interested in analysing data restricted to one country, or specific regions, instead of having a worldwide view, this document focuses on geographically segmenting data.

The following sections focuses on presenting the two international geocodes standards (NUTS, and GADM), followed by the required data-processing needed to achieve geographical segmentation.

<a id='Regional segmentation'></a>
### 1.1 Regional segmentation

#### NUTS

The **Nomenclature of Territorial Units for Statistics** (**NUTS**; French: Nomenclature des unités territoriales statistiques) is a geocode standard for referencing the subdivisions of countries for statistical purposes. The standard is developed and regulated by the European Union, and thus only covers the member states of the EU in detail [[2]](https://en.wikipedia.org/wiki/Nomenclature_of_Territorial_Units_for_Statistics).

For each EU member country, a hierarchy of three NUTS levels is established by [Eurostat](https://en.wikipedia.org/wiki/Eurostat) in agreement with each member state [[2](https://en.wikipedia.org/wiki/Nomenclature_of_Territorial_Units_for_Statistics)].

The subdivisions in some levels do not necessarily correspond to administrative divisions within the country [[2](https://en.wikipedia.org/wiki/Nomenclature_of_Territorial_Units_for_Statistics)]. In general, however, one can simplify and think of NUTS as follows:

* **NUTS 0**: country level

* **NUTS 1**: groups of states within a contry

* **NUTS 2**: regions within a contry

* **NUTS 3**: districts within a contry

Therefore, granularity increases with increasing NUTS' level.

For a better understanding, please refer to [https://en.wikipedia.org/wiki/Nomenclature_of_Territorial_Units_for_Statistics#Levels](https://en.wikipedia.org/wiki/Nomenclature_of_Territorial_Units_for_Statistics#Levels).

#### GADM

**GADM**, the **Database of Global Administrative Areas**, is a high-resolution database of country administrative areas, with a goal of mapping the administrative areas of all countries, at all levels of sub-division [[3](https://en.wikipedia.org/wiki/GADM), [4](https://gadm.org/about.html)]. GADM may be viewd as an extension of NUTS applied to the globe.

### 2.3. Map the shapefile onto the data objects
<a id='Functions to rasterize a shapefile onto a given xarray object'></a>
#### 2.3.1. Functions to rasterize a shapefile onto a given xarray object

```python
#Function provided by Stephan Hoyer
# url: https://github.com/pydata/xarray/issues/501#issuecomment-126461466

def transform_from_latlon(lat, lon):
    lat = np.asarray(lat)
    lon = np.asarray(lon)
    trans = Affine.translation(1.00111*lon[0], 1.00000*lat[0])  # transformation corrected
    scale = Affine.scale(lon[1] - lon[0], lat[1] - lat[0])
    return  trans * scale

def rasterize(shapes, coords, latitude='lat', longitude='lon',
              fill=np.nan, **kwargs):
    """Rasterize a list of (geometry, fill_value) tuples onto the given
    xray coordinates. This only works for 1d latitude and longitude
    arrays.
    """
    transform = transform_from_latlon(coords[latitude], coords[longitude])
    
    out_shape = (len(coords[latitude]), len(coords[longitude]))
    #requires the features library from rasterio
    raster = features.rasterize(shapes, out_shape=out_shape,
                                fill=fill, transform=transform, all_touched=True, dtype=float, **kwargs)
    spatial_coords = {latitude: coords[latitude], longitude: coords[longitude]}
    return xr.DataArray(raster, coords=spatial_coords, dims=(latitude, longitude))
```

<a id='Execute the rasterize function on FWI data'></a>
#### 2.3.2. Execute the rasterize function on FWI data

```python
%%time
# FWI data
shapes = zip(PT.geometry, range(len(PT)))
fwiPT = rasterize(shapes, ds.coords, longitude='lon', latitude='lat')
```

### 2.4. Segmentation

```python
ax = PT.plot(figsize=(20,5), alpha=0.8, color="black")
fwiPT.loc[dict(lat = slice(latBounds[0],latBounds[1]),
               lon = slice(lonBounds[0],lonBounds[1])
              )
         ].plot(ax=ax, alpha=0.5, zorder=-1)
```

```python
%%time
dsPT = ds.loc[dict(time = slice('1980-01-01', '2018-06-30'),
              lat = slice(latBounds[0],latBounds[1]), 
            lon = slice(lonBounds[0],lonBounds[1])
           )]
dsPT['fwiPT'] = fwiPT
```

```python
dsPT
```

```python
%%time
# applying a mask
newPT = dsPT.where(~np.isnan(dsPT.fwiPT)).drop('fwiPT')  # newPT:= segmented ds
```

```python
# plot
ax = PT.plot(figsize=(20,5), alpha=0.8, color="black")
newPT.fwi[0].plot(ax=ax, alpha=0.5, zorder=-1)
```
